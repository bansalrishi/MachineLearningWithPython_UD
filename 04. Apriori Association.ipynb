# Apriori

$Support \longrightarrow Confidence \longrightarrow Lift$  
  
Series Recommendation Engine: (S is set of 2 or series)  


### Support
$$Support(S) = \frac{No. of user watchlists contains S}{No. of user watchlists} $$
E.g: Out of 100 People 15 Watched GOT

$ Support = \frac {15}{100} = 15 \% $

### Confidence
$$Confidence(S_1 \longrightarrow S_2) = \frac{No. of user watchlists contains S_1 \$ S_2}{No. of user watchlists containing S_1} $$
E.g: Out of 100 People -> 50 Watched Prison & among these 50 -> 15 Watched GOT

$ Confidence = \frac {15}{50} = 30 \% $

### Lift
- Lift is improvement in the prediction  
$$Lift(S_1 \longrightarrow S_2) = \frac{Confidence (S_1 \longrightarrow S_2)}{Support (S_2)} $$
  
$ Lift = \frac {30\%}{15\%} = 2 $



### Algorithm

$ Step 1:$ Set Min. Support and Confidence  
$ Step 2:$ Take all the subsets in the transactions having higher Support than the min. support    
$ Step 3:$ Take all the rules of these subsets having higher confidence than min. confidence    
$ Step 4:$ Sort the rules by decreasing Lift  


```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from apyori import apriori

# Data Preprocessing
dataset = pd.read_csv(r'Data\Unsupervised\Market_Basket_Optimisation.csv', header = None)
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])

# Training Apriori on the dataset
association_rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)

# Storing the results
association_results = list(association_rules)

print(association_results[0])
```
